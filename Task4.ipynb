```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Create Dataset
data = {
    'Review': [
        'Excellent service',
        'Very good product',
        'Average experience',
        'Poor support',
        'Loved it',
        'Not satisfied',
        'Good quality',
        'Bad experience',
        'Okay service',
        'Fantastic product',
        'Fast delivery',
        'Terrible quality',
        'Average packaging',
        'Highly recommended',
        'Could be better'
    ],
    'Sentiment': [
        'Positive',
        'Positive',
        'Neutral',
        'Negative',
        'Positive',
        'Negative',
        'Positive',
        'Negative',
        'Neutral',
        'Positive',
        'Positive',
        'Negative',
        'Neutral',
        'Positive',
        'Neutral'
    ]
}

# Create DataFrame
df = pd.DataFrame(data)

# Save CSV File Automatically
df.to_csv('customer_feedback.csv', index=False)

print("CSV file saved successfully!")
print(df.head())

# Read CSV File
df = pd.read_csv('customer_feedback.csv')

# Count Sentiments
sentiment_count = df['Sentiment'].value_counts()

# Graph 1: Bar Chart
plt.figure(figsize=(6,4))
plt.bar(sentiment_count.index, sentiment_count.values)
plt.title('Sentiment Distribution')
plt.xlabel('Sentiment')
plt.ylabel('Count')
plt.show()
```

```python
# Graph 2: Pie Chart
plt.figure(figsize=(6,6))
plt.pie(sentiment_count.values,
        labels=sentiment_count.index,
        autopct='%1.1f%%')
plt.title('Sentiment Percentage')
plt.show()

# Graph 3: Count Plot
plt.figure(figsize=(6,4))
sns.countplot(data=df, x='Sentiment')
plt.title('Sentiment Count')
plt.show()

# Graph 4: Horizontal Bar Chart
plt.figure(figsize=(6,4))
sentiment_count.plot(kind='barh')
plt.title('Customer Feedback Sentiment Analysis')
plt.xlabel('Count')
plt.ylabel('Sentiment')
plt.show()

# Summary
print("\nSentiment Summary:")
print(sentiment_count)
```
